# Training with YOLOv8

```python
# Download pretrain model
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)
```

```python
"""
path ของ data.yaml สามารถหาได้จาก tap ด้านซ้ายที่เมนู file
data.yaml จะอยู่ใน folder ที่เรา download datasets จาก Roboflow
"""
path=""

# epochs คือจำนวนรอบในการ Train ปรับตามความเหมาะสม
results = model.train(data=path, epochs=10, imgsz=640)
```

# Inference custom trained model

```python
# Doc: https://docs.ultralytics.com/modes/predict/
from ultralytics import YOLO

# path ของ model ที่เรา train
custom_model = "/content/runs/detect/train2/weights/best.pt"

# Load a pretrained YOLO11n model
model = YOLO(custom_model)

# path ของรูปที่เราต้องการจะทำนาย
test_image = "/content/My-First-Garbage-Project--1/valid/images/000121_JPG.rf.c0cfbcb9e75f8768493fe956bdea376a.jpg"

model.predict(test_image, save=True, imgsz=640, conf=0.2)
```
